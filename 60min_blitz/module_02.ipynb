```python
import torch, torchvision
model = torchvision.models.resnet18(pretrained=True)
data = torch.rand(1, 3, 64, 64)
labels = torch.rand(1, 1000)
```

```python
print(f"Data:\n{data.shape}\n{data}")
print(f"Label:\n{labels}")
```

```python
# forward pass
prediction = model(data)
print(f"Prediction:\n{prediction.shape}\n{prediction}")
```

```python
err = prediction - labels
print(f"Error:\n{err.shape}\n{err}")
```

```python
loss = err.sum()
print(f"Loss:\n{type(loss)}\n{loss.shape}\n{loss}")
```

```python
#loss.backward()
print(loss.grad)
```

```python
optim = torch.optim.SGD(model.parameters(), lr=1e-2, momentum=0.9)
```

```python
optim.step() #gradient descent
```
